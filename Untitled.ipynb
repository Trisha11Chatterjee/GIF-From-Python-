```python
import matplotlib.pyplot as plt
import imageio
import os
import numpy as np

def create_frame(t):
    fig = plt.figure(figsize=(6, 6))
    plt.plot(x[:(t+1)], y[:(t+1)], color = 'grey' )
    plt.plot(x[t], y[t], color = 'black', marker = 'o' )
    plt.xlim([0,5])
    plt.xlabel('x', fontsize = 14)
    plt.ylim([0,5])
    plt.ylabel('y', fontsize = 14)
    plt.title(f'Relationship between x and y at step {t}',
              fontsize=14)
    plt.savefig(f'./img/img_{t}.png',
                transparent = False,
                facecolor = 'white'
               )
    plt.close()

# Sample data (replace with your actual data)
x = np.linspace(0, 5, 10)
y = x**2

# Number of frames
num_frames = len(x)

# Create img directory
os.makedirs('./img', exist_ok=True)

# Create and save individual frames
for t in range(num_frames):
    create_frame(t)

# Calculate grid dimensions for subplots
num_cols = int(np.ceil(np.sqrt(num_frames)))
num_rows = int(np.ceil(num_frames / num_cols))

# Create figure and subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 4, num_rows * 4))
fig.suptitle('All Timeframes', fontsize=16)

# Flatten axes array for easier iteration
axes = axes.flatten()

# Plot each frame on a subplot
for i, frame_num in enumerate(range(num_frames)):
    image_path = os.path.join('./img', f'img_{frame_num}.png')
    image = imageio.imread(image_path)

    axes[i].imshow(image)
    axes[i].axis('off')  # Hide subplot axes
    axes[i].set_title(f"Frame: {frame_num}")

# Remove any empty subplots (if num_frames is not a perfect square)
for j in range(num_frames, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
